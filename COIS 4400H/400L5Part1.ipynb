```python
#Part 1

#read in the Iris Data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
iris = datasets.load_iris()

print(iris) #prints 
```

```python
#Cluster the data using varying number of clusters
#K-means algorithm

from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

#creating dataframe
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names) 

#last column values excluded
x = iris_df.iloc[:, :-1].values 
#last column value
y = iris_df.iloc[:,   -1].values 

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)



#k-means 
kmeans = KMeans(n_clusters=5,init = 'k-means++',   max_iter = 100, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)

#display cluster centers
print(kmeans.cluster_centers_) 

#Visualising the clusters - On the first two columns
plt.scatter(x[y_kmeans   == 0, 0], x[y_kmeans == 0, 1],s = 100, label = 'Cluster1')
plt.scatter(x[y_kmeans   == 1, 0], x[y_kmeans == 1, 1],s = 100, label = 'Cluster2')
plt.scatter(x[y_kmeans   == 2, 0], x[y_kmeans == 2, 1],s = 100, label = 'Cluster3')
plt.scatter(x[y_kmeans   == 3, 0], x[y_kmeans == 3, 1],s = 100, label = 'Cluster4')
plt.scatter(x[y_kmeans   == 4, 0], x[y_kmeans == 4, 1],s = 100, label = 'Cluster5')
#plt.scatter(x[y_kmeans   == 5, 0], x[y_kmeans == 5, 1],s = 100, label = 'Cluster6')
#plt.scatter(x[y_kmeans   == 6, 0], x[y_kmeans == 6, 1],s = 100, label = 'Cluster7')
#plt.scatter(x[y_kmeans   == 7, 0], x[y_kmeans == 7, 1],s = 100, label = 'Cluster8')
#plt.scatter(x[y_kmeans   == 8, 0], x[y_kmeans == 8, 1],s = 100, label = 'Cluster9')
#plt.scatter(x[y_kmeans   == 9, 0], x[y_kmeans == 9, 1],s = 100, label = 'Cluster10')   

#scatter plot of clusters
#plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s = 100, c = 'black', label = 'Centroids')   
plt.legend()
plt.title("Iris Data k-means, Cluster = 5")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal.width")
plt.show()

```
