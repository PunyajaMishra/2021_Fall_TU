# **Part 2a : Transform any Categorical Attribute **

```python
#using panda 
#importing all other necessary libraries
#importing boston house pricing dataset
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

from sklearn import datasets
boston = datasets.load_boston()
```

```python
#Trying to UNDERSTAND the data before working with it
#data has a few keys
boston.keys()

#details of dataset
#data is the dataset
#target is the depndent variable: the prices 
#feature_names is the attributes/column names
#DECR is the detials about the dataset, columns and all
```

```python
boston.feature_names
#the column names
```

```python
print(boston.DESCR)
#details about dataset - we get to know what these columns names mean
#506 instances - rows
#14 attributes - columns
#no missing data
```

```python
#printing the dataset now
#reference (Still learning python and data analysis using python) -
#https://www.kaggle.com/jayateerthas/boston-dataset-analysis  

#convert the dataset into pandas dataframe
#setting the dfx values of df (dataframe) as the attributes
dfx = pd.DataFrame(boston.data, columns = boston.feature_names)
#setting the dfy values of df (dataframe) as the target/the actual data 
dfy = pd.DataFrame(boston.target,columns=['target'])

#now combining the 2
dfc = dfx.join(dfy)

#view and examine the head of the combined dataframe
pd.set_option('max_rows', 99999)
dfc

#from the dataset printed below we an see that RAD 
# 1, 2, 3, 5, 4, 8, 6, 7, 24 
#==> 1, 2, 3, 4, 5, 6, 7, 8, 24
#thus we can say that RAD is a categorical attribute
#low - 1, 2, 3
#medium - 4, 5
#high - 6, 7, 8
#very high - 24
```

```python
dfc['RAD']
```

```python
#RAD is a categorical attirbute
#first we need to make sure data knwos RAD is categorical
dfc.RAD = dfc.RAD.astype('category')
dfc.RAD = dfc.RAD.cat.reorder_categories([24.0,8.0,7.0,6.0,5.0,4.0,3.0,2.0,1.0])

replace_dict = {
    1.0:"Low",
    2.0:"Low",
    3.0:"Low",
    4.0:"Medium",
    5.0:"Medium",
    6.0:"High",
    7.0:"High",
    8.0:"High",
    24.0:"Very High"}

dfc.replace({"RAD":replace_dict}, inplace=True)

dfc.RAD
```

# **Part 2b : transform numerical target to 3 classes  using equal number of samples for each class**

```python
#continuous numerical values
#make 3 equal classes and divide all these values into the 3 samples
print(pd.cut(dfc.target,3, labels=[1,2,3]))
```

# ***# Part 3 : Decision Classifier Tree and Confusion Matrix ***

```python
from sklearn import preprocessing
from sklearn import utils

X = boston.data
y = boston.target
y = y.round().astype(int)
#had to convert target values from float to int otherwise it was giving me error 
#for the data to be continuous 
#Error : ValueError: Unknown label type: 'continuous'
X_trainingData, X_testData, y_trainingData, y_testData = train_test_split(X,y,test_size=0.33)


myClassifier=DecisionTreeClassifier()
myClassifier.fit(X_trainingData, y_trainingData)
prediction = myClassifier.predict(X_testData)

print(classification_report(y_testData, prediction))
print(confusion_matrix(y_testData, prediction))
```
