**First way : Scatter Plot**

```python
import matplotlib.pyplot as plt
```

```python
from sklearn import neighbors, datasets
iris = datasets.load_iris()
```

```python
x = 0 #sepal length col index
y = 1 #sepal width col index
```

```python
# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])

#set the output figure size inXin
plt.figure(figsize=(5, 4))
#scatter plot pf 2 features 0 and 1 (sepal length and width)
plt.scatter(iris.data[:, x], iris.data[:, y], c=iris.target)
#colorbar; tick locations to match desired ticklabels
plt.colorbar(ticks=[0, 1, 2], format=formatter)
#x_Axis label
plt.xlabel(iris.feature_names[x])
#y_Axis label
plt.ylabel(iris.feature_names[y])
#title
plt.title("Punyaja Mishra 0660001")

#tight layout automatically adjusts subplot parameters to give specfied padding
plt.tight_layout()

plt.show()
```

**Second way : Histogram**

```python
fig, axes = plt.subplots(nrows= 2, ncols=2)
colors= ['blue', 'red', 'green']

for i, ax in enumerate(axes.flat):
    for label, color in zip(range(len(iris.target_names)), colors):
        ax.hist(iris.data[iris.target==label, i], label=             
                            iris.target_names[label], color=color)
        ax.set_xlabel(iris.feature_names[i]) 
        ax.set_ylabel('Number') 
        ax.set_title("Punyaja Mishra 0660001")
        ax.legend(loc='upper right')
        

plt.show()
```
